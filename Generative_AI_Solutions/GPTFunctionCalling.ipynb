```python
import openai
from openai import OpenAI

api_key = "" # YOUR API KEY
api_base = "https://openai.vocareum.com/v1"

client = OpenAI(
    api_key=api_key,
    base_url=api_base
)
```

```python
# Define the function
def sample_function(input_string):
    return f"Function called with argument: {input_string}"
```

```python
# Define the tools
# This tools list contains the functions that the assistant can use.
# Each function is described by its name, description and parameters
tools = [
    {
        "type": "function",
        "function": {
            "name": "sample_function",
            "description": "A sample function that returns a string",
            "parameters":{
                "type": "object",
                "properties": {
                    "input_string": {
                        "type": "string",
                        "description": "A string to pass to the function"
                    }
                },
                "required": ["input_string"]
            }
        }
    }
]
```

```python
# Define the initial conversation messages
# The first message is from the system, describing the role of the assistant.
# The second message is from the user, requesting the function call.
messages = [
    {
        "role": "system",
        "content": "You are an assistant that can call a sample function."
    },
    {
        "role": "user",
        "content": "Call the sample function 10 times with the argument 'Hello there, World 'i'!, where i is the value ranges from 1 to 10'."
    }
]
```

```python
# Generate response from the model
response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages,
    tools=tools,
    tool_choice="auto",
)
```

```python
import json

# Extract the tool calls and call the function
# If there are any tool calls, map the tool names to their corresponding python functions.
# For each tool call in the response, extract the function name and call the corresponding python function with the arguments provided by the model

tools_calls = response.choices[0].message.tool_calls

if tools_calls:
    available_function = {
        "sample_function": sample_function,
    }
    
    for tool_call in tools_calls:
        function_name = tool_call.function.name
        function_to_call = available_function[function_name]
        function_args = json.loads(tool_call.function.arguments)
        print(function_args)
        if function_name == 'sample_function':
            function_response = function_to_call(
                input_string=function_args.get("input_string"),
            )
        
        messages.append({
            "role":"assistant",
            "content":function_response
        })
        
```

```python
for message in messages:
    print(f"{message['role']}: {message['content']}")
```
