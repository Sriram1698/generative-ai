```python
# Objective: Develop a prompt template that guides the LLM to generate a concise, sentiment-focused summary
# based on user inputs: the restaurant name and the type of cuisine.
```

```python
# Create variables to store the user inputs
restaurant_name = "Alinea"
cuisine_type = "new american"
```

```python
# Create the prompt template
prompt_template = f"Provide a summary of customer sentiments for {restaurant_name}, focusing on their {cuisine_type} dishes. Highlight key sentiments and mention any standout dishes or services. "
print(prompt_template)
```

```python
import openai
from openai import OpenAI

api_key = "" # YOUR API KEY
api_base = "https://openai.vocareum.com/v1"

client = OpenAI(
    api_key=api_key,
    base_url=api_base
)
```

```python
# Function to call the OpenAI GPT-3.5-Turbo API
def generate_restaurant_review(prompt_template):
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {
                    "role": "system",
                    "content": "You are a restaurant critic. You write detailed, balanced reviews of restaurants."
                },
                {
                    "role": "user",
                    "content": prompt_template
                }
            ],
            temperature=1,
            max_tokens=256,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )
        # Extract the assistant's reply
        return response.choices[0].message.content
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# Generating the response from the model
review_summary = generate_restaurant_review(prompt_template)

# Printing the output. 
print("Generated review:")
print(review_summary)
```
